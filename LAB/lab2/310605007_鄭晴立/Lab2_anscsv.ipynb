```python
import shutil
import os
 
file_source = './data/testing/all/'
file_destination = './data/test_ans_new/'
 

# shutil.copyfile(file_source , file_destination)
    
    
for i in range(len(dataset_test)):
    for j in range(len(dataset)):
        if dataset_test.imgs[i][0][19:]==dataset.imgs[j][0][32:] and dataset.imgs[j][0][19:31]=="glioma_tumor":
            print(dataset_test.imgs[i][0][19:],"||",dataset.imgs[j][0][32:])
            shutil.copyfile(file_source+str(dataset_test.imgs[i][0][19:]) , file_destination+str(dataset.imgs[j][0][19:31])+'/'+str(dataset_test.imgs[i][0][19:]))
       
        elif dataset_test.imgs[i][0][19:]==dataset.imgs[j][0][36:] and dataset.imgs[j][0][19:35]=="meningioma_tumor":
            print(dataset_test.imgs[i][0][19:],"||",dataset.imgs[j][0][36:])
            shutil.copyfile(file_source+str(dataset_test.imgs[i][0][19:]) , file_destination+str(dataset.imgs[j][0][19:35])+'/'+str(dataset_test.imgs[i][0][19:]))
        
        elif dataset_test.imgs[i][0][19:]==dataset.imgs[j][0][28:] and dataset.imgs[j][0][19:27]=="no_tumor":
            print(dataset_test.imgs[i][0][19:],"||",dataset.imgs[j][0][28:])
            shutil.copyfile(file_source+str(dataset_test.imgs[i][0][19:]) , file_destination+str(dataset.imgs[j][0][19:27])+'/'+str(dataset_test.imgs[i][0][19:]))

        
        elif dataset_test.imgs[i][0][19:]==dataset.imgs[j][0][35:] and dataset.imgs[j][0][19:34]=="pituitary_tumor":
            print(dataset_test.imgs[i][0][19:],"||",dataset.imgs[j][0][35:])
            shutil.copyfile(file_source+str(dataset_test.imgs[i][0][19:]) , file_destination+str(dataset.imgs[j][0][19:34])+'/'+str(dataset_test.imgs[i][0][19:]))
```

```python
with open ('ID_result_high.csv','w') as f:
    f.write('ID,label\n')
    for i in range(len(dataset)):
        if dataset.imgs[i][0][19:31]=="glioma_tumor":
            f.write(dataset.imgs[i][0][32:]+','+str(dataset.imgs[i][1])+'\n')
        elif dataset.imgs[i][0][19:35]=="meningioma_tumor": 
            f.write(dataset.imgs[i][0][36:]+','+str(dataset.imgs[i][1])+'\n')
        elif dataset.imgs[i][0][19:27]=="no_tumor": 
            f.write(dataset.imgs[i][0][28:]+','+str(dataset.imgs[i][1])+'\n')
        elif dataset.imgs[i][0][19:34]=="pituitary_tumor": 
            f.write(dataset.imgs[i][0][35:]+','+str(dataset.imgs[i][1])+'\n')


```
